```python
# define test variables
outputDirectory = '/data/home/jenhan/analysis/ap1_analysis/output_mergeTest/veh'
genome = 'mm10'
size = 4000 
peakFilePaths = ['/data/home/jenhan/analysis/ap1_analysis/workingPeaks_mergeTest/veh/cjun-notx_peaks.tsv',
'/data/home/jenhan/analysis/ap1_analysis/workingPeaks_mergeTest/veh/gro-notx_peaks.tsv',
'/data/home/jenhan/analysis/ap1_analysis/workingPeaks_mergeTest/veh/h3k4me1-notx_peaks.tsv',
'/data/home/jenhan/analysis/ap1_analysis/workingPeaks_mergeTest/veh/pu1-notx_peaks.tsv']
distFile = '/data/home/jenhan/analysis/ap1_analysis/output_mergeTest/veh/distance.tsv'
sameStrand = False
makeVisuals = True
makePeakFile = True
```

```python
# run script - essentially main method of script

# if out put directory does not exist, create it
if not os.path.exists(outputDirectory):
    os.makedirs(outputDirectory)

# read peak files and get tag directories
peakDataFrames = [] # list of data frames that hold peak data

revMergeDistDict = {} # key: expA|expB, value: sorted list of distances
                      # for merging peaks on the reverse strand 
fwdMergeDistDict = {} # key: expA|expB, value: sorted list of distances
                      # for merging peaks on the forward strand 

for peakFilePath in peakFilePaths: 
    peakdf = readPeakFile(peakFilePath)
    peakDataFrames.append(peakdf)
if not distFile == None:
    distanceDicts = readDistanceFile(distFile)
    revMergeDistDict = distanceDicts[0]
    fwdMergeDistDict = distanceDicts[1]
else:
    # write merge distances for editing
    distanceFile = open(outputDirectory + "/distance.tsv", "w")
    distanceFile.write("\t".join(["name1", "name2", "fwdDistance", 
                                  "revDistance"]) +"\n")
    # call Homer to compute tag densities for each pair of peak files
    for i in range(len(peakDataFrames) - 1): 
        for j in range(i+1, len(peakDataFrames)):
            tagDir1 = peakDataFrames[i].tagDirectory
            filePath1 = peakDataFrames[i].filePath
            name1 = peakDataFrames[i].name
            style1 = peakDataFrames[i].style.lower()

            tagDir2 = peakDataFrames[j].tagDirectory
            filePath2 = peakDataFrames[j].filePath
            name2 = peakDataFrames[j].name
            style2 = peakDataFrames[j].style.lower()

            # center around non-groseq file if possible
            if (style1 == "groseq" and not style2 == "groseq"):
                # switch name style pairs
                print("GRO-Seq experiment detected, centering around" + \
                       " non-GRO-Seq experiment: " + name2,
                       file = sys.stderr)
                tempName = name1
                tempStyle = style1
                tempTagDir = tagDir1
                tempFilePath = filePath1

                name1 = name2
                style1 = style2
                tagDir1 = tagDir2
                filePath1 = filePath2

                name2 = tempName
                style2 = tempStyle
                tagDir2 = tempTagDir
                filePath2 = tempFilePath

            # create tag density file; center around first peak file
            histFileName = outputDirectory + "/" + name1 + "_" + name2 \
                           + "_hist.txt"

            if not os.path.isfile(histFileName):
                print("Homer is calculating tag densities for: " + name1 +
                      " " + name2, file = sys.stderr)
                os.system("annotatePeaks.pl " + filePath1 + " " + genome +
                    " -hist 10 -size " + str(size) + " -d " + tagDir1 + " "
                    + tagDir2 + " > " + histFileName)
            # read in tag density file and compute optimal distance
            distanceTuple = calcTagDensityDistance(histFileName,
                                                        peakDataFrames[i],
                                                        peakDataFrames[j],
                                                        makeVisuals)

            # TODO create test for whether or not peaks used are any good
            if len(distanceTuple[0]) == 0 and len(distanceTuple[1]) == 0:
                # first orientation does not give an optimal distance; 
                #center around the second peak file
                print("Centering around " + name1 +
                      " did not produce optimal results - " +
                     "trying to center around " + name2, file = sys.stderr)

                histFileName2 = outputDirectory + "/" + name2 + "_" + \
                                name1 + "_hist.txt"
                print("Using: " + histFileName2, file = sys.stderr)

                if not os.path.isfile(histFileName2):
                    print("Homer is calculating tag densities for: " +name2+
                          " " + name1, file = sys.stderr)
                    os.system("annotatePeaks.pl " + filePath2 + " " + genome
                        + " -hist 10 -size " + str(size) + " -d " + tagDir2
                        + " " + tagDir1 + " > " + histFileName2)
                distanceTuple2 = calcTagDensityDistance(histFileName2,
                                                        peakDataFrames[j],
                                                        peakDataFrames[i],
                                                        makeVisuals)
                # TODO find a better measure; for now use tuple with more values
                if len(distanceTuple2[0]) + len(distanceTuple2[1]) > \
                   len(distanceTuple[0]) + len(distanceTuple[1]):
                    distanceTuple = distanceTuple2

            if not distanceTuple[0] and not distanceTuple[1]:
                # both orientations does not give an optimal distance
                # avoid merging these two kinds of peaks unless there 
                # is literal overlap 
                print("Can't find a good merge distance for " + name1 +
                    " and " + name2, file = sys.stderr)
                print("Peaks from these peak files won't be merged",
                      file = sys.stderr)
            #incorporate distance values into 
            revMergeDistDict[name1+"|"+name2] = distanceTuple[0]
            revMergeDistDict[name2+"|"+name1] = distanceTuple[0]

            fwdMergeDistDict[name1+"|"+name2] = distanceTuple[1]
            fwdMergeDistDict[name2+"|"+name1] = distanceTuple[1]
            # write pair into distanceFile
            if distanceTuple[0]:
                fwdDistString = ",".join([str(x) for x in distanceTuple[0]])
            else:
                fwdDistString = "-"
            if distanceTuple[1]:
                revDistString = ",".join([str(x) for x in distanceTuple[1]])
            else:
                revDistString = "-"
            distanceFile.write("\t".join([name1, name2, fwdDistString,
                                revDistString]) + "\n")
    distanceFile.close()

# merge peaks together
network = mergePeaks(peakDataFrames, revMergeDistDict, fwdMergeDistDict,
                     sameStrand)
# write network as pickle file
network.write_pickle(outputDirectory + "/network_pickle")

if makePeakFile:
    writePeakFile(network, outputDirectory + "/network_merged.tsv")                   
```

```python
### reshape homer_mergedPeaks_frame so it looks like network_mergedPeaks_frame ###
# delete columns that aren't needed
homer_mergedPeaks_frame = homer_mergedPeaks_rawFrame.copy()
homer_mergedPeaks_frame[homer_mergedPeaks_frame.columns[8:]]

# generate combined peak IDs
cols = list(homer_mergedPeaks_frame.columns[8:].values)
originalIDs = []
for index, row in homer_mergedPeaks_frame[homer_mergedPeaks_frame.columns[8:]].fillna('').iterrows():
    peakIDs = list(row.values)
    combinedID = []
    for i in range(len(cols)):
        if peakIDs[i]:
            tokens = peakIDs[i].split(",")
            for t in tokens:
                combinedID.append(cols[i]+'_'+t)
    originalIDs.append('|'.join(combinedID))

# delete extra columns
for delCol in homer_mergedPeaks_frame.columns[7:]:
    del homer_mergedPeaks_frame[delCol]
del homer_mergedPeaks_frame["Stat"]
homer_mergedPeaks_frame.columns = ['PeakID', 'chr', 'start', 'end', 'strand', 'parent files']

# add original peak IDs to homer data frame
homer_mergedPeaks_frame['original IDs'] = originalIDs   

```
