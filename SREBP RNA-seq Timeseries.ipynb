# Knockout 

## Read in Data

```python
data_frame = pd.read_csv('SREBP_mm9_rpkm.txt', sep='\t')
columns = data_frame.columns.values
columns[0] = 'refseq'
columns = [col.replace('reads FPKM','').replace('notx','0h').strip() for col in columns]
data_frame.columns = columns
data_frame.index = data_frame['refseq'].values
value_frame = data_frame.ix[:,8:]
# work only with genes that are expressed in at least one sample
value_frame = value_frame[value_frame.sum(axis=1) > 1]
print('Number of genes expressed in at least 1 sample', value_frame.shape[0])

refseq_gene_mapping = dict(zip(data_frame['refseq'].values,
                               [x.split('|')[0] for x in data_frame['Annotation/Divergence'].values]))
```

```python
log_frame = np.log2(value_frame).replace([np.inf, -np.inf], 0)
sns.violinplot(log_frame)
plt.xticks(rotation=90)
plt.ylabel('Log2 RPKM')
plt.xlabel('Sample Name')
plt.title('Distribution of Log RPKM values')
```

```python
wt_cols = sorted([x for x in value_frame.columns.values if 'WT' in x ], key = lambda x: int(x.split('_')[-1][:-1]))

ko_cols = sorted([x for x in value_frame.columns.values if 'KO' in x ], key = lambda x: int(x.split('_')[-1][:-1]))
```

```python
difference_dict = {}
for i in range(len(wt_cols)):
    wt_col = wt_cols[i]
    ko_col = ko_cols[i]
    differences = value_frame[wt_col] - value_frame[ko_col]
    signs = [-1 if x < 0 else 1 for x in differences]
    differences = signs * np.log2(np.abs(differences)).replace([np.inf, -np.inf], 0)
    difference_dict[wt_col.split('_')[-1]] = differences
difference_frame = pd.DataFrame(difference_dict)

# sort the columns for the plot
difference_frame = difference_frame[sorted(difference_frame.columns.values, key=lambda x: int(x[:-1]))]

sns.violinplot(difference_frame)

plt.ylabel('log(WT-KO) RPKM')
plt.xlabel('Timepoint')
plt.title('Log Distribution of Differences Between KO and WT')
```

```python
# find significant differentially expressed genes
xlabels = ['0h','1h', '6h','12h', '24h','48h' ]
genes = [refseq_gene_mapping[x] for x in value_frame.index.values]
significant_genes_knockout = []
rpkms = value_frame.index.values
for i in range(value_frame.shape[0]):
    gene = genes[i]
    wt_values = value_frame.ix[i,wt_cols].values.flatten()
    ko_values = value_frame.ix[i,ko_cols].values.flatten()
    stat, p = scipy.stats.ks_2samp(wt_values, ko_values)
    if p< 0.05:
        print(gene)
        significant_genes_knockout.append(rpkms[i])
#         plt.plot(range(len(wt_values)), wt_values)
#         plt.plot(range(len(ko_values)), ko_values)
#         plt.legend(['Control', 'Knockdown'])
#         plt.xticks(range(len(wt_values)), ['0h','1h', '6h','12h', '24h','48h'])
#         plt.xlabel('Time')
#         plt.ylabel('RPKM')
#         plt.title(gene + ' pval= ' + str(p))
#         plt.savefig(gene+'_sigDifference.png')
#         plt.close()

```

```python
sns.clustermap(np.log2(value_frame.ix[significant_genes_knockout] + 0.0000001), col_cluster= False , yticklabels=False)
```

```python
data_frame_knockdown = pd.read_csv('srebp_siRNA_rpkm.tsv', sep='\t')
columns = data_frame_knockdown.columns.values
columns[0] = 'refseq'
columns = [col.replace('reads FPKM','').replace('notx','0h').strip() for col in columns]
data_frame_knockdown.columns = columns
data_frame_knockdown.index = data_frame_knockdown['refseq'].values
value_frame_knockdown = data_frame_knockdown.ix[:,8:]
# work only with genes that are expressed in at least one sample
value_frame_knockdown = value_frame_knockdown[value_frame_knockdown.sum(axis=1) > 1]
print('Number of genes expressed in at least 1 sample', value_frame_knockdown.shape[0])
refseq_gene_mapping = dict(zip(data_frame_knockdown['refseq'].values,
                               [x.split('|')[0] for x in data_frame_knockdown['Annotation/Divergence'].values]))
```

```python
wt_cols = sorted([x for x in value_frame_knockdown.columns.values if 'sicont' in x ],
                 key = lambda x: int(x.split('_')[-1][:-1]))

ko_cols = sorted([x for x in value_frame_knockdown.columns.values if 'siSREBF1' in x ], 
                 key = lambda x: int(x.split('_')[-1][:-1]))

```

```python
difference_dict = {}
for i in range(len(wt_cols)):
    wt_col = wt_cols[i]
    ko_col = ko_cols[i]
    differences = value_frame_knockdown[wt_col] - value_frame_knockdown[ko_col]
    signs = [-1 if x < 0 else 1 for x in differences]
    differences = signs * np.log2(np.abs(differences)).replace([np.inf, -np.inf], 0)
    difference_dict[wt_col.split('_')[-1]] = differences
difference_frame = pd.DataFrame(difference_dict)

# sort the columns for the plot
difference_frame = difference_frame[sorted(difference_frame.columns.values, key=lambda x: int(x[:-1]))]

sns.violinplot(difference_frame)

plt.ylabel('log(WT-KO) RPKM')
plt.xlabel('Timepoint')
plt.title('Log Distribution of Differences Between KO and WT')
```

```python
# find significant differentially expressed genes
xlabels = ['0h','3h', '6h','24h' ]
significant_genes_knockdown = []
genes = [refseq_gene_mapping[x] for x in value_frame_knockdown.index.values]
rpkms = value_frame_knockdown.index.values
for i in range(value_frame_knockdown.shape[0]):
    gene = genes[i]
    wt_values = value_frame_knockdown.ix[i,wt_cols].values.flatten()
    ko_values = value_frame_knockdown.ix[i,ko_cols].values.flatten()
    stat, p = scipy.stats.ks_2samp(wt_values, ko_values)
    if p< 0.05:
        significant_genes_knockdown.append(rpkms[i])
#         plt.plot(range(len(wt_values)), wt_values)
#         plt.plot(range(len(ko_values)), ko_values)
#         plt.legend(['Control', 'Knockdown'])
#         plt.xticks(range(len(wt_values)), ['0h','3h', '6h','24h' ])
#         plt.xlabel('Time')
#         plt.ylabel('RPKM')
#         plt.title(gene + ' pval= ' + str(p))
#         plt.savefig(gene+'_sigDifference.png')
#         plt.close()
```
