# Specific Aim 2: How does AP-1 contribute to enhancer activity?

```python
def readTransitionMatrix(transitionMatrixPath):
    ''' 
    Reads in a transition matrix formated as tsv file, returning a list of states and a Numpy matrix
    inputs: path to a rtransition matrix represented as a tsv file
    outputs: a list of states, the transition matrix as a numpy matrix
    '''
    with open(transitionMatrixPath) as f:
        data = f.readlines()
    states = data[0].strip().split("\t")
    transitionMatrix = np.zeros((len(states), len(states)))
    for i in range(len(data) - 1): 
        tokens = data[i + 1].strip().split("\t")[1:]
        for j in range(len(tokens)):
            transitionMatrix[i][j] = float(tokens[j])
    
    # normalize using row totals
    for i in range(len(states)):
        transitionMatrix[i] = transitionMatrix[i]/np.sum(transitionMatrix[i])
    

    return states, transitionMatrix

# calculates the Pearson correlation between two motifs
# inputs: two motifs
# outputs: a Pearson correlation value
def calcCorrelation(motif1, motif2):
        a = 0.0 
        b = 0.0 
        c = 0.0 
        for i in range(motif1.shape[0]):
                for j in range(len(motif1[0])):
                        dif1 = motif1[i][j] - np.mean(motif1[i])  
                        dif2 = motif2[i][j] - np.mean(motif2[i])
                        a += dif1 * dif2
                        b += dif1 ** 2.0 
                        c += dif2 ** 2.0 
        toReturn = a/((b*c) ** 0.5)
        return toReturn
```

```python
# compare transition matrices
import seaborn as sns
import os
matricesDir = "/data/home/jenhan/analysis/eventOrder_analysis/output_mol_cell_noh3k4me1/unnormalized_transitionMatrices/"
paths = os.listdir(matricesDir)
matrices = []
matrixNames = []
for path in paths:
    if "tsv" in path:
        states, matrix = readTransitionMatrix(matricesDir + path)
        matrices.append(matrix)
        name = "-".join([x.split("-")[1] for x in path.replace(".tsv","").split("_")])
        matrixNames.append(name)
        
# calc correlations
corrMatrix = np.zeros((len(matrices), len(matrices)))

for i in range(len(matrices)-1):
    for j in range(i+1, len(matrices)):
        corr = calcCorrelation(matrices[i], matrices[j])
        corrMatrix[i][j] = corr
        corrMatrix[j][i] = corr
corrMatrixFrame = pd.DataFrame(corrMatrix)
corrMatrixFrame.index = matrixNames
corrMatrixFrame.columns = matrixNames
sns.heatmap(corrMatrixFrame, annot = True, square = True)
```
