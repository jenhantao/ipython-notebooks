```python
columns = diffExpression_frame.columns.values
columns[0] = 'Refseq'
columns[7] = 'Gene'
diffExpression_frame.columns = columns
diffExpression_frame = diffExpression_frame[['Refseq', 'Gene'] + [x for x in columns if 'vs.' in x]]
diffExpression_frame['Gene'] = [x.split('|')[0] for x in rpkm_frame['Gene'].values]
```

```python
kla_inducible_genes = diffExpression_frame[(diffExpression_frame['WT_Veh vs. WT_KLA adj. p-value'] < 0.05) &
                                           (diffExpression_frame['WT_Veh vs. WT_KLA Log2 Fold Change'] >=2)]['Gene'].values
```

```python
sns.clustermap(np.log2(rpkm_mean_frame[rpkm_mean_frame['Gene'].isin(kla_inducible_genes)][['KO_KLA', 'WT_KLA+DHA', 'WT_KLA+EPA']]+1),
               yticklabels=False,
              vmax=4)
```

```python
threshold = 4
expressed_genes = rpkm_mean_frame[rpkm_mean_frame.ix[:,['WT_KLA', 'WT_KLA+EPA', 'WT_KLA+DHA', 'KO_KLA']].max(axis=1) >= threshold]['Gene'].values
```

```python
pseudocount=1
data = (rpkm_mean_frame[(rpkm_mean_frame['Gene'].isin(kla_inducible_genes)) &
                        (rpkm_mean_frame['Gene'].isin(expressed_genes))][['KO_KLA', 'WT_KLA+DHA', 'WT_KLA+EPA']]+pseudocount).div(
 (rpkm_mean_frame[(rpkm_mean_frame['Gene'].isin(kla_inducible_genes)) &
                   (rpkm_mean_frame['Gene'].isin(expressed_genes))]['WT_KLA']+pseudocount),
 axis=0)
log_data = np.log2(data)
```
